```python
import cv2 as cv
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
for i in range(1, 10):
    file_path = "/home/jachinshen/Downloads/fireNumber/{}.jpg".format(i)
    img = cv.imread(file_path)
    img = cv.cvtColor(img, cv.COLOR_BGR2GRAY)
    img = img[:,int(img.shape[1] * 0.2) : int(img.shape[1] * 0.8)]
    img = cv.GaussianBlur(img, (51, 51), 0)
    _, maxVal, _, _ = cv.minMaxLoc(img)
    _, img = cv.threshold(img, 0.80 * maxVal, 255, cv.THRESH_BINARY)
    print(maxVal)
    img = cv.resize(img, (56, 56))
    plt.imshow(img)
    cv.imwrite("{}.jpg".format(i), img)
    plt.show()
```
